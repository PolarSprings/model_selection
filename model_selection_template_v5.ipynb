```python
from sklearn.preprocessing import OneHotEncoder

def OneHotPotato(df, catcol):
    df = df.reset_index().rename(columns={'index':'id'})
    df2 = pd.DataFrame(OneHotEncoder().fit_transform(df[[str(catcol)]]).toarray(),
                                  columns = [str(catcol) + '_' + str(i) for i in np.unique(df[str(catcol)].values)])

    df2['id'] = range(len(df2))
    
    return df.drop([catcol],1).merge(df2, on='id').drop(['id'],1)
```

```python
col_lst = ['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','Churn']

for col in col_lst:
    churn = OneHotPotato(churn, col)
```

### balancing the data with SMOTE

```python
from imblearn.over_sampling import SMOTE

def SMOTE_oversample(df, dep_var):

    data = df.drop([str(dep_var)],1).squeeze()
    target = df[[str(dep_var)]].squeeze()

    oversample = SMOTE()

    data, target = oversample.fit_resample(data, target)
    
    return data, target

data, target = SMOTE_oversample(churn, 'Churn_Yes')
```

### model scoring: train, test, cv, precision, recall, accuracy, f1

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import BayesianRidge
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVR
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
import lightgbm as lgb

models_lst = [LinearRegression(), 
          ElasticNet(), 
          BayesianRidge(), 
          MultinomialNB(), 
          SVR(),
          GaussianNB(),
          DecisionTreeClassifier(),
          RandomForestClassifier(),
          AdaBoostClassifier(),
          GradientBoostingClassifier(), 
          KNeighborsClassifier(), 
          LogisticRegression(solver='liblinear'), 
          xgb.XGBClassifier(),
          lgb.LGBMClassifier()]
```

```python
%%time

from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

def model_selection_kfold(models, data, target): 

    idx = np.random.choice(range(0,len(data)), size=round(len(data)*0.80), replace=False)

    x_train = data.iloc[idx].reset_index(drop=True)
    y_train = target.iloc[idx].reset_index(drop=True)

    x_validate = data.iloc[data.index.difference(idx)].reset_index(drop=True)
    y_validate = target.iloc[data.index.difference(idx)].reset_index(drop=True)  
    
    idx = np.random.choice(range(0,len(x_train)), size=round(len(x_train)*0.50), replace=False)

    x_test = x_train.iloc[x_train.index.difference(idx)].reset_index(drop=True)
    y_test = y_train.iloc[x_train.index.difference(idx)].reset_index(drop=True)
    
    x_train = x_train.iloc[idx].reset_index(drop=True)
    y_train = y_train.iloc[idx].reset_index(drop=True)
    
    
    scores_dct = {}    
    
    for model in models:

        clf = model

        clf.fit(x_train, y_train)

        preds = clf.predict(x_test)

        mod_auc = roc_auc_score(y_test, preds)
        mod_fpr, mod_tpr, i = roc_curve(y_test, preds)

        cv_score = np.mean(cross_val_score(clf, data, target, cv=KFold(n_splits=5, shuffle=True, random_state=42)))

        scores_dct[model] = {'precision':metrics.precision_score(preds.round(), y_test), 
                             'recall':metrics.recall_score(preds.round(), y_test), 
                             'accuracy':metrics.accuracy_score(preds.round(), y_test), 
                             'f1':metrics.f1_score(preds.round(), y_test),
                             'train':clf.score(x_train, y_train),
                             'test':clf.score(x_test, y_test),
                              'cv':cv_score,
                              'auc':mod_auc,
                              'fpr':mod_fpr,
                              'tpr':mod_tpr}


#         print('\n')
#         print('The model', model, 'had the following Classification Report')
#         print(metrics.classification_report(preds,y_test))
#         print('-----------------------------------------------------------')
#         print('\n')
    
    return scores_dct, x_train, y_train, x_test, y_test, x_validate, y_validate, preds

scores_dct, x_train, y_train, x_test, y_test, x_validate, y_validate, preds = model_selection_kfold(models_lst, data, target)
```

```python
def scores_dct_to_df(scores_dictionary):

    scores_df = pd.DataFrame(columns=[str(name).replace('()','') for name in models_lst],
                            index=['train','test','precision','recall','accuracy','f1','auc','fpr','tpr'])

    scores_df.columns = [("XGBClassifier") if "XGB" in col else col for col in scores_df]
    
    for score in ['train','test','cv','precision','recall','accuracy','f1','auc']:
        scores_df.loc[str(score), ::] = [np.mean(i[str(score)]) for i in scores_dictionary.values()]

    for score in ['fpr','tpr']:
        scores_df.loc[str(score), ::] = [i[str(score)] for i in scores_dictionary.values()]
        
    return scores_df

scores_kf_df = scores_dct_to_df(scores_dct)
scores_kf_df
```

```python
%%time

from sklearn.model_selection import GridSearchCV

clf = AdaBoostClassifier()

data = churn.drop(['Churn_Yes'],1).squeeze()
target = churn[['Churn_Yes']].squeeze()


gs = GridSearchCV(
    clf,
    param_grid={
        'n_estimators':[1000,5000,10000],
        'learning_rate':[0.01,0.1,0.5],  
        'base_estimator':[DecisionTreeClassifier(max_depth=trees) for trees in range(1,11)]
    },
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
)

gs.fit(data, target)
```
